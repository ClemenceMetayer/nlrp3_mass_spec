# ---------------------- Importation of the data and results ----------------------<a class="anchor" id="importation_data"></a>

```python
df_data_mass_spec = pd.read_csv("Mass Spec/cosinor_data_concat_serie.csv")
df_results_mass_spec = pd.read_csv("Mass Spec/cosinor_results_concat_serie.csv")
df_results_mass_spec = df_results_mass_spec.loc[df_results_mass_spec['period']==24.2,]

df_data_rna_seq = pd.read_csv("RNA-Seq/cosinor_data_concat_serie.csv")
df_results_rna_seq = pd.read_csv("RNA-Seq/cosinor_results_concat_serie.csv")
df_results_rna_seq = df_results_rna_seq.loc[df_results_rna_seq['period']==24.2,]

df_rain_data_rna_seq = pd.read_csv("RNA-Seq/rain_results.csv")
df_rain_data_mass_spec = pd.read_csv("Mass Spec/rain_results.csv")
```

# ----------------------------- Circadian clock results -----------------------------<a class="anchor" id="circadian_clock"></a>

## <font color=#31B404> RNA-Seq data <a class="anchor" id="rna_seq_data"></a></font>

### CLOCK <a class="anchor" id="clock"></a>

```python
clock_data = df_data_rna_seq.loc[df_data_rna_seq["test"]=="CLOCK/ctrl",]
clock_results = df_results_rna_seq.loc[df_results_rna_seq['test']=="CLOCK/ctrl",]
clock_pval = float(clock_results["p"])
clock_qval = float(clock_results["q"])
print(f"For the CLOCK transcript, the Cosinor model with a 24.2h period has a {clock_pval} p-value and a {clock_qval} q-value.")
```

```python
cosinor.plot_df_models(clock_data, clock_results)
```
